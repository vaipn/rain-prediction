```python
# Function to find missing values in the dataset
def find_missing_values(df):
    missing_values = []
    for date in df['Date'].unique():
        date_df = df[df['Date'] == date]
        for location in date_df['Location'].unique():
            loc_df = date_df[date_df['Location'] == location]
            if loc_df.isnull().values.any():
                for index, row in loc_df.iterrows():
                    if row.isnull().any():
                        missing_values.append({'Date': date,'Location': location,'Index': index,'Missing Values': row.isnull().values})
    return pd.DataFrame(missing_values)
```

```python
# Function to fill missing values
def fill_missing_values(df, missing_values, city_distances):
    for index, row in missing_values.iterrows():
        location = row['Location']
        date = row['Date']
        missing_columns = np.where(row["Missing Values"])[0]
        #print("missing columns: ", missing_columns)
        nearest_city = city_distances[location]
        #print("nearest city", nearest_city)
        nearest_city_data = df[(df['Date'] == row['Date']) & (df['Location'] == nearest_city)]
        #print("nearest_city_data: ", nearest_city_data)
        print("Row: ", row)
        if not nearest_city_data.empty:
            for column in missing_columns:
                #print("Missing column: ", column)
                #print("Original Dataframe row", df.loc[index, column])
                #print("Replacing DF row", nearest_city_data[column].iloc[0])
                #print(nearest_city_data.iloc[0, column])
                df.iloc[index, column] = nearest_city_data.iloc[0, column]
        else:
            nearest_cities = list(df[df['Location'] != location]['Location'].unique())
            #print("Nearest Cities: ", nearest_cities)
            nearest_cities_distances = {}
            for city in nearest_cities:
                nearest_cities_distances[city] = geodesic(city_coordinates[location], city_coordinates[city]).km
            nearest_cities_distances = dict(sorted(nearest_cities_distances.items(), key=lambda item: item[1]))
            four_cities = list(nearest_cities_distances.keys())[:5]
            #print("5 nearest cities: ", four_cities)
            four_cities_data = df[(df['Date'] == row['Date']) & (df['Location'].isin(four_cities))]
            if len(four_cities_data) >= 4:
                for column in missing_columns:
                    if df.iloc[:, column].dtype == 'object':
                        continue
                    df.iloc[index, column] = four_cities_data.iloc[:, column].mean()
            else:
                week_data = df[(df['Date'] >= date - timedelta(days=7)) & (df['Date'] <= date + timedelta(days=7))]
                #print("Week Data: ", week_data)
                for column in missing_columns:
                    #print("Missing column: ", column)
                    #print("Original Dataframe column value", df.loc[row['Index'], column])
                    if df.iloc[:, column].dtype == 'object':
                        continue
                    #print("Replacing column", week_data.iloc[:, column])
                    df.iloc[index, column] = week_data.iloc[:, column].mean()   
    return df

```
